# Histogram Graphs

Code for Histogram graphical presentation

Descriptive statistics are important for a brief overview of the dataset we are using. One way to display this in an easier to understand format is with histograms. Histograms are partiuclarly useful for displaying the distribution of a given dataset. Since this dataset is concerned with test scores, knowing the dsitribution will be useful for any further statsitical testing. 

First, we will import our dataset and libraries.

```python
# %matplotlib inline
# pyplot should make the above line redundant
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0, 10.0)

df = pd.read_csv('exams.csv')
```

While calcualtions would correctly show the distribution of our varialbes, a histogram will be much more intuitive

First, we will show the distribution of the math scores

```python
#Plot histogram of math scores
df['math score'].plot(kind='hist')
```

```python
#adjust bins and display of histogram
fig, ax = plt.subplots(figsize=(10,5))
plt.hist(df['math score'],bins=25)
```

To further list our median, mode, and standard deviation, we will calculate it here first

```python
#Calculate mean of math scores
df[['math score']].mean()
```

```python
df[['math score']].median()
```

```python
df[['math score']].mode()
```

```python
#calculate standard deviation of math scores
df[['math score']].std()
```

```python
#importing different library not covered in class
import seaborn as sns
from scipy import stats
from scipy.optimize import curve_fit
from scipy import asarray as ar,exp

from scipy.optimize import curve_fit
```

```python
data = df['math score'].values

# Plotting the histogram
plt.hist(data, bins=20, density=True, alpha=0.6, color='b')

plt.show()
```

We can see here that the histogram appears to have a unimodal distribution, but we will graph it to be sure.

```python
from scipy.stats import norm

# Fit a normal distribution to the data:
# mean and standard deviation
mu, std = norm.fit(data)

# Plotting the histogram.
plt.hist(data, bins=20, density=True, alpha=0.6, color='b')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2)
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)

plt.show()
```

As we can see here, the math scores have a normal distribution, suggesting that most observations can be found around the mean.
